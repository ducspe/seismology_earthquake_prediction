```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.HDFStore("../../cloudcontainer/concatenated_experiments.hdf5", mode='r').select("alldata_mean_instead_of_NaN")

df_train_minmax = pd.HDFStore("../../cloudcontainer/concatenated_experiments.hdf5", mode='r').select("alldata_mean_instead_of_NaN/train/minmax")
df_val_minmax = pd.HDFStore("../../cloudcontainer/concatenated_experiments.hdf5", mode='r').select("alldata_mean_instead_of_NaN/val/minmax")
df_test_minmax = pd.HDFStore("../../cloudcontainer/concatenated_experiments.hdf5", mode='r').select("alldata_mean_instead_of_NaN/test/minmax")

df_train_zscale = pd.HDFStore("../../cloudcontainer/concatenated_experiments.hdf5", mode='r').select("alldata_mean_instead_of_NaN/train/zscale")
df_val_zscale = pd.HDFStore("../../cloudcontainer/concatenated_experiments.hdf5", mode='r').select("alldata_mean_instead_of_NaN/val/zscale")
df_test_zscale = pd.HDFStore("../../cloudcontainer/concatenated_experiments.hdf5", mode='r').select("alldata_mean_instead_of_NaN/test/zscale")

print(df.shape)
print(df_train_minmax.shape)
print(df_val_minmax.shape)
print(df_test_minmax.shape)
```

```python
pressure = df["pressure"]
pressure_train_minmax = df_train_minmax["pressure"]
pressure_val_minmax = df_val_minmax["pressure"]
pressure_test_minmax = df_test_minmax["pressure"]

pressure_train_zscale = df_train_zscale["pressure"]
pressure_val_zscale = df_val_zscale["pressure"]
pressure_test_zscale = df_test_zscale["pressure"]

ae = df["ae"]
ae_train_minmax = df_train_minmax["ae"]
ae_val_minmax = df_val_minmax["ae"]
ae_test_minmax = df_test_minmax["ae"]

ae_train_zscale = df_train_zscale["ae"]
ae_val_zscale = df_val_zscale["ae"]
ae_test_zscale = df_test_zscale["ae"]
```

```python
train_len = len(df_train_minmax)
val_len = len(df_val_minmax)
test_len = len(df_test_minmax)
print(train_len, val_len, test_len)
```

```python
fig = plt.figure(figsize=(30,12))
ax1 = fig.add_subplot(131)
plt.plot(pressure)
plt.title("Pressure plot")
plt.ylabel('Pressure')
plt.xlabel('Time step')
plt.axvline(x=14400, color='r')
plt.axvline(x=14400+17515, color='r')
#plt.axvline(x=train_len, color='g')
#plt.axvline(x=train_len+val_len, color='black')

ax2 = fig.add_subplot(132)
plt.plot(pressure_train_minmax)
plt.plot(pressure_val_minmax)
plt.plot(pressure_test_minmax)
plt.title("Pressure minmax plot")
plt.ylabel('Pressure minmax')
plt.xlabel('Time step')
plt.axvline(x=14400, color='r')
plt.axvline(x=14400+17515, color='r')
#plt.axvline(x=train_len, color='g')
#plt.axvline(x=train_len+val_len, color='black')

ax3 = fig.add_subplot(133)
plt.plot(pressure_train_zscale)
plt.plot(pressure_val_zscale)
plt.plot(pressure_test_zscale)
plt.title("Pressure Z-scale plot")
plt.ylabel('Pressure Z-scale')
plt.xlabel('Time step')
plt.axvline(x=14400, color='r')
plt.axvline(x=14400+17515, color='r')
#plt.axvline(x=train_len, color='g')
#plt.axvline(x=train_len+val_len, color='black')
```
